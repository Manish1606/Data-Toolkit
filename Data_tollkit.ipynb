Q2. How does broadcasting work in NumPy?
==> It works by "stretching" or "replicating" smaller arrays to
   match the dimensions of larger arrays.
it works:-
  * NumPy compares the shapes of the arrays from the rightmost dimension and works its way left.
  * If the arrays have different numbers of dimensions, NumPy adds a "wrapper dimension" to the shape of the lower-rank array.
  * If the shapes of the arrays don't match in any dimension, NumPy stretches the array with shape 1 in that dimension to match the other shape.
  * If the sizes disagree in any dimension and neither is equal to 1, NumPy raises an error.

```python
#Q5. Generate a bar plot using Plotly.

import plotly.graph_objects as go
import plotly.express as px
import pandas as pd

# Load the iris dataset
df = px.data.iris()

# Display the first five rows of the data
print(df.head())

# Create bar plot
bar_trace = go.Bar(
    x=df.species,
    y=df.petal_length,
    name='Petal'
)

# Create figure and add trace
fig = go.Figure(bar_trace)

# Customize layout
fig.update_layout(title='Bar Plot of Petal Length by Species')

# Display interactive plot
fig.show()
```
